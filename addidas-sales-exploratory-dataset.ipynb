```python
#PRINTING OUT MY INVOICE DATE COLUMNS FROM 0 TO 50


import pandas as pd
import numpy as np
import datetime as dt
df = pd.read_csv('/kaggle/input/adidas/data_sales (1).csv')
df['Invoice Date'].unique()[:50]
```

```python
#CONVERTING MY DATE INTO A TO DATE TIME FORMAT


df['Invoice Date'] = pd.to_datetime(df['Invoice Date'])
df['Invoice Date']
```

```python

#EXTRACTING THE MONTH IN INVOICE DATE


df['Month'] = df['Invoice Date'].dt.month
df['Month']
```

```python
#EXTRACTING THE YEAR IN INVOICE DATE


df['Year'] = df['Invoice Date'].dt.year
df['Year']
```

```python
#EXTRACTING THE DAY IN INVOICE DATE


df['Day'] = df['Invoice Date'].dt.day 
df['Day']
```

```python
#CREATING A SEASON COLUMNS

def find_seasons(MonthNumber):
    if MonthNumber in [12, 1, 2]:
        return 'winter'
    
    elif MonthNumber in [3, 4, 5]:
        return 'spring'
    
    elif MonthNumber in [6, 7, 8]:
        return 'summer'
    
    elif MonthNumber in [9, 10, 11]:
        return 'autumn'
    
df['Seasons'] = df['Month'].apply(find_seasons)
df['Seasons']
```

```python
#CONVERTING THE MONTH COLUME INTO MONTH NAMES


df['Month'] = pd.to_datetime(df['Month'], format = '%m').dt.month_name()
```

```python
#DROPPING THE RETAILERID AND INVOICE DATE COLUMNS


df.drop(columns = ['Retailer ID', 'Invoice Date'], inplace = True)
```

```python
#REPLACING SOME COLUMNS DATA TYPES 

df['Day'] = df['Day'].astype('category')
df['Seasons'] = df['Seasons'].astype('category')
df['Year'] = df['Year'].astype('category')
```

```python
#REMOVING THE DOLLAR SIGN WITH THE FUNCTION I CREATED BEFORE


df['Operating Profit'] = df['Operating Profit'].apply(removeDollarSign)
df['Operating Profit']
```
